```python
from preprocessing.preprocessing_pipeline.preprocessing_pipeline import PreprocessingPipeline
import pandas as pd
```

```python
path = "../../../data/Trainings_Datensatz.csv"

pipeline = PreprocessingPipeline(
    path=path,
    basic_features=True,
    data_augmentation=False,
    calculate_indices=True,
    temporal_features=True,
    interpolate_b4=True,
    outlier_cleaner=False,
    detect_disturbed_trees=False, 
    specify_disturbed_labels=True,
)

df_cleaned = pipeline.run()

df_cleaned.to_csv("../../../data/processed/df_processed.csv", index=False)
# TO DO:
# sin columns as flag 
```
