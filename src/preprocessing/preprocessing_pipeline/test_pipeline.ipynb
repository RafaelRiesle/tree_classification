```python
from preprocessing.preprocessing_pipeline.preprocessing_pipeline import PreprocessingPipeline
```

```python
path = "../../../data/Trainings_Datensatz.csv"

pipeline = PreprocessingPipeline(
    path=path,
    basic_features=True,
    calculate_indices=True,
    temporal_features=True,
    interpolate_b4=True,
    outlier_cleaner=False
)

df_cleaned = pipeline.run()
```

```python
df_cleaned
```

```python
# test class detect disturbed trees
from preprocessing.preprocessing_steps.detect_disturbed_trees import DetectDisturbedTrees

detector = DetectDisturbedTrees(scale_pos_weight=4, random_state=42)
results_df = detector.run(df_cleaned)
```

```python
results_df
```

```python
print(len(results_df[results_df["is_disturbed_pred"] == 0]))
print(len(results_df[results_df["is_disturbed_pred"] == 1]))

# with scale_pos_weight = 5: 
# 22044
# 2538
```

```python
# validate model result

top_features = ["b11_slope", "b5_slope", "b11_std", "gndvi_std"]
results_df["combi_top_features"] = results_df[top_features].mean(axis=1)

results_df[results_df["is_disturbed_pred"] == 0].sort_values(by="combi_top_features", ascending=False)
```
